```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('evaluation.csv')
data= data.sort_values(by='Accuracy')
```

```python
data_sorted = data.sort_values(by='Accuracy')
plt.figure(figsize=(8, 6))
bars = plt.bar(data_sorted['Model'], data_sorted['Accuracy'], color='skyblue', edgecolor='black', linewidth=0.8,width=0.4)

plt.xlabel('Modelo')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy de los Modelos de Análisis de Sentimiento')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, round(yval, 2), va='bottom', ha='center', color='black')

plt.ylim(60, 90)


plt.tight_layout()
plt.show()
```

```python
data_sorted = data.sort_values(by='Accuracy',ascending=False)
plt.figure(figsize=(12, 6)) 
bars = plt.barh(data_sorted['Model'], data_sorted['Accuracy'], color='skyblue', edgecolor='black', linewidth=0.8,height=0.4)

plt.xlabel('Accuracy (%)')
plt.ylabel('Modelo')
plt.title('Accuracy de los Modelos de Análisis de Sentimiento')

for bar in bars:
    xval = bar.get_width()
    plt.text(xval + 0.2, bar.get_y() + bar.get_height()/2, round(xval, 2), va='center', ha='left', color='black')

plt.xlim(70, 90)
```

```python
colors = ['skyblue', 'orange', 'crimson']

barWidth = 0.25
plt.figure(figsize=(10, 6))

r1 = np.arange(len(data['Model']))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

for i, metrica in enumerate(['F1', 'Precision', 'Recall']):
    plt.bar(eval(f'r{i+1}'), data[metrica], color=colors[i], width=barWidth, edgecolor='black', label=metrica)

plt.xlabel('Modelo', fontweight='bold')
plt.ylabel('Porcentaje', fontweight='bold')
plt.xticks([r + 1 * barWidth for r in range(len(data['Model']))], data['Model'])
plt.title('COmparación de Metricas de los Modelos de Análisis de Sentimiento')

plt.legend()

for r1_val, r2_val, r3_val, index in zip(r1, r2, r3, range(len(data['Model']))):
    modelo = data.iloc[index]['Model']
    plt.text(r1_val , data.iloc[index]['F1'] + 1, f"{data.iloc[index]['F1']}%", ha='center')
    plt.text(r2_val , data.iloc[index]['Precision'] + 1, f"{data.iloc[index]['Precision']}%", ha='center')
    plt.text(r3_val , data.iloc[index]['Recall'] + 1, f"{data.iloc[index]['Recall']}%", ha='center')

metrica_patch = [plt.Rectangle((0, 0), 1, 1, color=color) for color in colors]
plt.legend(metrica_patch, ['F1', 'Precision', 'Recall'], loc='upper right')

plt.ylim(30, 100)

plt.tight_layout()
plt.show()
```
